```python
df["Average_Marks"] = (df["Math_Marks"] + df["Python_Marks"]) / 2
```

```python
sns.barplot(data=df,x="Student",y="Python_Marks")
plt.title("Python Marks for Students")
plt.xlabel("Students")
plt.ylabel("Python_Marks")
plt.show()
```

```python
sns.barplot(
data=df,
x="Student",
y="Python_Marks"
)
plt.title("Python Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Python Marks")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df,x="Department",y="Python_Marks")
plt.title("Average Python Marks of Departments")
plt.xlabel("Department Name")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.countplot(data=df,x="Department")
plt.title("Count of students in Departments")
plt.xlabel("Department Name")
plt.ylabel("Number of Students")
plt.show()
```

```python
sns.histplot(data=df,x="Python_Marks")
plt.title("Histogram of Python_Marks")
plt.xlabel("Python_Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(
data=df,
x="Python_Marks",
kde=True
)
plt.title("Distribution of Python Marks(The curved line is KDE)")
plt.xlabel("Python Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df,x="Python_Marks",kde=True,bins=5)
plt.title("Distribution of python marks")
plt.show()
```

```python
sns.boxplot(data=df,x="Department",y="Python_Marks")
plt.title("Box plot of python marks")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.boxplot(data=df,x="Department",y="Python_Marks")
plt.title("Box plot of python marks")
plt.xlabel("Department Name")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(data=df,x="Attendance",y="Python_Marks")
plt.title("Scatter plot of attendance")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(data=df,x="Attendance",y="Python_Marks",hue="Department")
plt.title("Scatter plot of attendance")
plt.xlabel("Attendance")
plt.ylabel("Python Marks")
plt.show()
```

```python
sns.scatterplot(data=df,x="Attendance",y="Python_Marks",hue="Department",size="Average_Marks")
plt.title("Scatter plot of attendance")
plt.xlabel("Attendance")
```
